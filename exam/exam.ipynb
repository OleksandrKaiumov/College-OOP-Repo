## Реалізація

```python
class TemperatureConverter:
    def __init__(self, celsius: float):
        self.celsius = celsius

    def convert_to_fahrenheit(self) -> float:
        return self.celsius * 9 / 5 + 32

    @property
    def fahrenheit(self) -> float:
        return round(self.convert_to_fahrenheit(), 1)


# Ініціалізація об'єктів з різними значеннями температури
t1 = TemperatureConverter(0)
t2 = TemperatureConverter(25)
t3 = TemperatureConverter(-10)

print(f"Температура {t1.celsius}°C дорівнює {t1.fahrenheit}°F.")
print(f"Температура {t2.celsius}°C дорівнює {t2.fahrenheit}°F.")
print(f"Температура {t3.celsius}°C дорівнює {t3.fahrenheit}°F.")
```
