```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return amount
        else:
            return "Insufficient funds"

    def get_balance(self):
        return self.__balance

account = BankAccount("Bohdan", 1000)

count = 3
minNum = 100
maxNum = 500
for _ in range(count):
    account.deposit(random.randint(minNum, maxNum))
    account.withdraw(random.randint(minNum, maxNum))

print("Баланс: ", account.get_balance())
```

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"
    
    def display_brand(self):
        return f"Brand: {self.brand}"

class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)  # виклик конструктора базового класу
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Seats: {self.seats}"

car = Car("Toyota", "Camry", 5)
print(car.display_info())
print(car.display_brand())
```

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"
    
class Fish(Animal):
    pass

# Використання:
animals = [Dog(), Cat(), Fish()]
for animal in animals:
    print(animal.speak())
```

Метод speak класу Fish повернене результат методу speak класу Animal, тобто None, оскільки Fish успадковує і не перезаписує метод батьківського класу.

```python
from abc import ABC, abstractmethod
from random import randint, choice

class Item(ABC):
    def __init__(self, name:str, health = 500):
        self.name = name
        self.health = health

    def enhance(self):
        pass
    
    @abstractmethod
    def attack(self):
        pass

class Sword(Item): # Це наслідування
    def __init__(self, name, attack_power:int):
        super().__init__(name=name) #  Це виклик конструктора наслідуваного класу
        self.__attack_power = attack_power # приватний атрибут який реалізує інкапсуляцію
        self._sharp = 0
    
    def attack(self, another_item:Item): # ми не можемо створити клас без цього методу
        current_attack = self.__attack_power + self._sharp + randint(0, 10)
        another_item.health -= current_attack
        return f"Завдаємо удару мечем {self.name} та наносимо {current_attack} шкоди. У {another_item.name} залишалось здоров'я: {another_item.health}"
    
    @property
    def get_attack_power(self):
        return f"Атака меча {self.name}: {self.__attack_power + self._sharp} одиниць"
    
    def enhance(self):
        self._sharp += 1
        return f"Меч {self.name} заточено. Атака: {self.__attack_power + self._sharp} одиниць"

class Axe(Item): # Це наслідування
    def __init__(self, name, attack_power:int):
        super().__init__(name=name) #  Це виклик конструктора наслідуваного класу
        self.__attack_power = attack_power # приватний атрибут який реалізує інкапсуляцію
        self._sharp = 0
    
    def attack(self, another_item:Item):  # ми не можемо створити клас без цього методу
        current_attack = self.__attack_power + randint(0, 20)
        another_item.health -= current_attack
        return f"Завдаємо удару сокирою {self.name} та наносимо {current_attack} шкоди. У {another_item.name} залишалось здоров'я: {another_item.health}"

    @property
    def get_attack_power(self):
        return f"Атака сокири {self.name}: {self.__attack_power + self._sharp} одиниць"
    
class Bow(Item):
    def __init__(self, name, attack_power:int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._range_power = 0
    
    def attack(self, another_item:Item):
        current_attack = self.__attack_power + randint(5, 15) + self._range_power
        another_item.health -= current_attack
        return f"Завдаємо удару стрілою {self.name} та наносимо {current_attack} шкоди. У {another_item.name} залишалось здоров'я: {another_item.health}"

    def enhance(self):
        self._range_power += 1
        return f"Лук {self.name} покращено. Атака: {self.__attack_power + self._range_power} одиниць"

    @property
    def get_attack_power(self):
        return f"Атака луку {self.name}: {self.__attack_power + self._range_power} одиниць"

def get_random_item():
    weapons = [
        Sword("Ескалібур", 100),
        Axe("Кратос", 95),
        Bow("Артеміда", 90)
    ]
    return choice(weapons)
    
player_item = get_random_item()
enemy_item = get_random_item()

enemy_item.name += " (Суперник)"

turn = 1
while player_item.health > 0 and enemy_item.health > 0:
    print(f"Хід {turn}")
    print(f"Ваше здоров'я: {player_item.health}")
    print(f"Здоров'я суперника: {enemy_item.health}")
    print("Оберіть дію:")
    print("1. Атакувати")
    print("2. Підсилити зброю")

    action = 0
    while action < 1 or action > 2:
        action = int(input("Ваш вибір: "))
        if action == 1:
            print(f"\n[Гравець] {player_item.attack(enemy_item)}")
        elif action == 2:
            print(f"\n[Гравець] {player_item.enhance()}")
    
    if enemy_item.health <= 0:
            print(f"\nВи перемогли!")
            break

    action = randint(1, 2)
    if action == 1:
        print(f"[Суперник] {enemy_item.attack(player_item)}")
    else:
        print(f"[Суперник] {enemy_item.enhance()}")


    if player_item.health <= 0:
        print(f"\nПоразка.")
        break

    turn += 1
```
